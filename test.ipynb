```python
import numpy as np
import cv2,os,datetime,calendar
import pandas as pd

known_encodings = np.load(r"F:\face attendance system - Copy\known_faces_database\known_face_encodings.npy")  # load stored face-encodings
known_labels = np.load(r"F:\face attendance system - Copy\known_faces_database\known_face_labels.npy")  # load stored labels

names_list = known_labels.tolist()
names_list=[i.title() for i in names_list]
threshold_in_time=datetime.datetime.strptime("14:59:00",'%H:%M:%S')
```

```python
def base_df(names_list):
    """ Takes Names list ; creates all columns for that month-csv """
    today=datetime.datetime.today()                                                              # today
    current_month_yr=today.strftime("%b-%Y")
    
    out_csv_name=current_month_yr+".csv"   
    csv_out_path=os.path.join(r"F:\face attendance system - Copy\database\master_csv",out_csv_name)
    
    day,row,num_days=1,[],calendar.monthrange(today.year, today.month)[1]
    
    for i in range(num_days):
        d = "0" + str(day) if len(str(day)) == 1 else str(day)  # if 1:'01'; 17:'17'
        row.append(d + "-" + current_month_yr)
        day += 1

    """ generates static df with name & day-wise column(default 0) """
    df = pd.DataFrame(0, columns=names_list, index=row)
    df.to_csv(csv_out_path, index=True)
    print("CSV created @ ",csv_out_path)
    
#####
#base_df(names_list)
```
